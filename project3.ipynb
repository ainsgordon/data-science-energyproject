```python
#Brown Hall
brown = {}
browndates = []
brownkwh = []

for file in files:
    if file>= start and file<=end:
        total_kWh = 0
        f=open(file, 'r')
        for line in f:
            fields = line.split(',')
            if(building in fields[1]):
                total_kWh = total_kWh + float(fields[6])
        brown[file]=total_kWh
        f.close()

print("Electricity usage for", building)
print("Date \t kWh usage")
for date in sorted(brown.keys()):
    fmt_usage = "{:.2f}".format(brown[date])
    browndates.append(date[:8])
    brownkwh.append(fmt_usage)
    #print(date[:8], "\t", fmt_usage)

brown = {1: building, 2: browndates, 3: brownkwh}
```
